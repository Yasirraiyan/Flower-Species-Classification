```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# ডেটাসেট তৈরি করা
data = {
    'Petal Length': [1.4, 1.4, 4.7, 4.5, 5.1, 5.9],
    'Petal Width': [0.2, 0.2, 1.4, 1.5, 1.8, 2.1],
    'Species': ['Setosa', 'Setosa', 'Versicolor', 'Versicolor', 'Virginica', 'Virginica']
}
df = pd.DataFrame(data)

# ইনপুট এবং আউটপুট নির্ধারণ
X = df[['Petal Length', 'Petal Width']]
y = df['Species']

# কেএনএন ক্লাসিফায়ার তৈরি করা
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# নতুন ডেটা পয়েন্টের প্রজাতি নির্ধারণ
new_data = np.array([[4.8, 1.6]])
prediction = knn.predict(new_data)
print(f'Prediction for new data point (Petal Length 4.8, Petal Width 1.6): {prediction[0]}')

# ডেটা প্লট করা
plt.scatter(df['Petal Length'], df['Petal Width'], c=df['Species'].apply(lambda x: {'Setosa':0, 'Versicolor':1, 'Virginica':2}[x]), cmap='viridis')
plt.scatter(new_data[:, 0], new_data[:, 1], c='red', marker='X', s=100)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Flower Species Classification with KNN')
plt.show()
```
